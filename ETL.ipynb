```python
df1['release_date'] = pd.to_datetime(df1['release_date'], errors='coerce')
```

```python
def extract_year(text):
    match = re.search(r'\b\d{4}\b', str(text))
    if match:
        return int(match.group())
    return None
```

```python
df1['release_year'] = df1['release_date'].apply(extract_year)
```

```python
df1.dropna(subset=['release_year'], inplace=True)
```

```python
df1.dropna(subset=['genres'], inplace=True)
```

```python
df1.isnull().sum()
```

```python
def genres_in_year(year):
    #repito procedimiento anterior para filtrar
    df_filtered = df1[df1['release_date'].dt.year == year]

    all_genres = df_filtered['genres'].tolist()
    all_genres_flat = [genre for sublist in all_genres if isinstance(sublist, list) for genre in sublist]
    genre_counts = pd.Series(all_genres_flat).value_counts(sort=True, ascending=False)

    #cambio el parametro para que me muestre los 5 más repetidos.
    most_common_genres = genre_counts.index[:5]

    return [genre for genre in most_common_genres]

genres_in_year(2007)
```
